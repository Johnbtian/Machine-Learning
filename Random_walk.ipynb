```python
#plan: to write random walk fxn of nblocks; simple version
import random
def random_walk(n):
    """"Return cordinates after 'n' block random walk:"""
    #initializing the walk
    x=0
    y=0
  # Take a random walk with n steps using a for loop 
#use choice function to choose a step
    for i in range(n):
        step=random.choice(['N','S','E','W'])
        if step=='N':
            y=y+1
            
        elif step=="S":
            y=y-1
            
        elif step=="E":
            x=x+1
            
        else:
            x=x-1
            
    return (x,y) #return cordinates as a tuple

#testing the random walk: take 25 random  walks each 10 blocks long

for i in range(25):
    walk=random_walk(10)
#for each walk display coordinates & distance from home(=sum of absolute values of the x&y cordinates~sign is for dirrection)
    print(walk, "Distance from home=",abs(walk[0])+abs(walk[1]))

```

```python
#Monte Carlo: what is the longest random walk you can take so that on average you will end up 4 blocks or fewer from home?
import random
def random_walk2(n):
    """Return cordinates after 'n' block random walk."""
    x,y=0,0
    if i in range(n):
        (dx,dy)=random.choice([(0,1),(0,-1),(1,0),(-1,0)])
        x+=dx
        y+=dy
    return (x,y)

#monte carlo implementation
#conducting thousands of random trials & compute the % of those in a short walk home.
number_of_walks= 20000

#estimate th probability you will walk home for blocks of length 1 to 31 blocks
for walk_length in range(1, 31):
    no_transport = 0 #No of walks 4 or fewer blocks from home
#Monte Carlo loop of 10000 samples
    for i in range(number_of_walks):
        (x,y)=random_walk2(walk_length) #getting a walk of distance walk length
        distance= abs(x)+abs(y) #computing distance from home
        if distance<=4: #if distance from hom is <=4 increment the no_transport counter
            no_transport +=1
#computing the %of random walks that ended with the walk home
    no_transport_percentage= float(no_transport)/ number_of_walks
    print("Walk size = ", walk_length,"/% of no transport =", 100*no_transport_percentage)
    
```
