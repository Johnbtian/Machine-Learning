```python
import distutils.dir_util
import os
import os.path as path


# A python script to copy files and folders from your one place to another
# The month remote folder is replaced by your local month folders with all its contents
# Replace the list of courts with yours
#local path- where the data is, remote where you are copying to


def main():
    local_base_path = r'C:\Users\Admin\Downloads\TEMPLATE 3\TEMPLATE 3'
    remote_base_path = r'E:\DCRT HIGH COURT\DCRT HIGH COURT\INPUT\TEMPLATE 3'
    start_year = "2023"
   #end_year = "2022"
    fy_2022_2023 = {"02": start_year}

    list_of_courts = ["Bomet High Court", "Bungoma High Court", "Busia High Court", "Chuka High Court", "Eldoret High Court",
                      "Embu High Court", "Garissa High Court","Garsen High Court", "Homabay High Court", "Kabarnet High Court", "Kajiado High Court",
                      "Kakamega High Court","Kapenguria High Court", "Kericho High Court",
                      "Kerugoya High Court","Kiambu High Court", "Kisii High Court", "Kisumu High Court",
                      "Kitale High Court", "Kitui High Court", "Lodwar High Court", "Machakos High Court", "Makueni High Court", "Malindi High Court",
                      "Marsabit High Court", "Meru High Court", "Migori High Court", "Milimani AntiCorruption & Economic Crimes Division",
                      "Milimani Civil Division", "Milimani Commercial & Tax Division", "Milimani Constitution Division", "Milimani Criminal Division", "Milimani Family Division", "Milimani Judicial Review Division", "Mombasa High Court", "Muranga High Court",
                      "Naivasha High Court", "Nakuru High Court", "Nanyuki High Court", "Narok High Court", "Nyamira High Court",
                      "Nyandarua High Court", "Nyeri High Court",
                      "Siaya High Court", "Vihiga High Court", "Voi High Court"]
    for court in list_of_courts:
        for month, year in fy_2022_2023.items():
            local_file_path = os.path.join(local_base_path, court, year, month)
            remote_file_path = os.path.join(remote_base_path, court, year, month)
            # Copy file to another directory
            if path.exists(remote_file_path):
                if path.exists(local_file_path):
                    distutils.dir_util.copy_tree(local_file_path, remote_file_path)
                    print("copied", local_file_path, "to", remote_file_path)
                else:
                    print("Missing", local_file_path)
                print("Completed copying files")
            else:
                print("Remote Folders don't exist")
                break


if __name__ == '__main__':
    main()
```
